```python

def fill_missing_numeric(dataset2):
    numeric_cols = dataset2.select_dtypes(include=['number']).columns.tolist()
    for col in numeric_cols:
        if dataset2[col].isnull().any():  # Check if there are missing values in the column
            if dataset2[col].isnull().sum() / len(dataset2[col]) < 0.05:
                # If less than 5% of values are missing, replace with mean
                dataset2[col].fillna(dataset2[col].mean(), inplace=True)
            elif dataset2[col].nunique() > 10:
                # If the column has more than 10 unique values, replace with median
                dataset2[col].fillna(dataset2[col].median(), inplace=True)
            else:
                # Otherwise, replace with mode
                dataset2[col].fillna(dataset2[col].mode()[0], inplace=True)
    return dataset2

```

```python
dataset2 = fill_missing_numeric(dataset2)
```

```python
def convert_categorical_to_numeric(dataset2):
    for col in dataset2.select_dtypes(include='object').columns:
        # Sort the unique categories alphabetically and factorize
        sorted_categories = sorted(dataset2[col].unique())
        mapping = {category: i for i, category in enumerate(sorted_categories)}
        dataset2[col] = dataset2[col].map(mapping)
    return dataset2
```

```python
# Convert categorical columns to numeric
dataset2 = convert_categorical_to_numeric(dataset2)
```
