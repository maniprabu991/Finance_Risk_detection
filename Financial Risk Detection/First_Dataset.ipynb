```python

def fill_missing_numeric(dataset1):
    numeric_cols = dataset1.select_dtypes(include=['number']).columns.tolist()
    for col in numeric_cols:
        if dataset1[col].isnull().any():  # Check if there are missing values in the column
            if dataset1[col].isnull().sum() / len(dataset1[col]) < 0.05:
                # If less than 5% of values are missing, replace with mean
                dataset1[col].fillna(dataset1[col].mean(), inplace=True)
            elif dataset1[col].nunique() > 10:
                # If the column has more than 10 unique values, replace with median
                dataset1[col].fillna(dataset1[col].median(), inplace=True)
            else:
                # Otherwise, replace with mode
                dataset1[col].fillna(dataset1[col].mode()[0], inplace=True)
    return dataset1

```

```python
dataset1 = fill_missing_numeric(dataset1)
```

```python
def convert_categorical_to_numeric(dataset1):
    for col in dataset1.select_dtypes(include='object').columns:
        # Sort the unique categories alphabetically and factorize
        sorted_categories = sorted(dataset1[col].unique())
        mapping = {category: i for i, category in enumerate(sorted_categories)}
        dataset1[col] = dataset1[col].map(mapping)
    return dataset1
```

```python
# Convert categorical columns to numeric
dataset1 = convert_categorical_to_numeric(dataset1)
```
